```python
import pandas as pd
```

```python
data = pd.read_csv("train_triplets_sample.txt",header = None
                       ,delimiter=r"\s+")
  
# adding column headings
data.columns = ['User', 'Song', 'Play_Count']
  
# store dataframe into csv file
data.to_csv('train_triplets_sample_updated.csv', 
                index = None)
```

```python
df=pd.read_csv("train_triplets_sample_updated.csv")
```

```python
df.head()
```

```python
lst=df.values.tolist()
lst.sort()
```

```python
users=df["User"].tolist()
```

```python
unique_users=list(set(users))
```

```python
songs=df["Song"].tolist()
unique_songs=list(set(songs))
```

```python
unique_users_play_count = []
for i in unique_users:
    x=0
    count=0
    for j in lst:
        if(j[0]==i):
            x=1
            count+=j[2]
        
        elif(j[0]!=i and x==1):
            break
        
    unique_users_play_count.append(count)

```

```python
unique_users_play_count
```

```python
unique_songs_play_count = []
for i in unique_songs:
    x=0
    count=0
    for j in lst:
        if(j[1]==i):
            x=1
            count+=j[2]
        
        elif(j[1]!=i and x==1):
            break
        
    unique_songs_play_count.append(count)

```

```python
unique_songs_play_count
```

```python
play_count=df["Play_Count"].tolist()
unique_play_count=list(set(play_count))
x=max(unique_play_count)
print(x)
```

```python
final_lst = []

for j in lst:
    if(j[2]==x):
        final_lst.append(j)
```

```python
final_lst
```

```python
print(len(lst))
```

```python
print(len(final_lst))
```

```python
# to create final list as a dataframe and save it as a different.csv file

df_final = pd.DataFrame(final_lst, columns = ['User', 'Song', 'Max_Play_Count'])
```
